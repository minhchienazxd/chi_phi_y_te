**3. phát hiện sử lý giá trị ngoại lai**

```python
#hàm tính toán các ngoại lai
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python
# Loại bỏ ngoại lai
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Giữ lại các giá trị không phải ngoại lai
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
numeric_columns = ['price', 'milage', 'model_year']
# Phát hiện ngoại lai
for col in numeric_columns:
    outliers = detect_outliers(data, col)
    print(f'Ngoại lai của {col}:')
    print(outliers)
```

```python
# Loại bỏ ngoại lai
for col in numeric_columns:
    data = remove_outliers(data, col)
```

```python
# Kiểm tra lại dữ liệu sau khi loại bỏ ngoại lai
print(f'Dữ liệu sau khi xử lý ngoại lai:\n{data.describe()}')
```
