```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
dfs = sns.load_dataset("penguins",cache=False)
```

```python
dfs.head()
```

```python
df = dfs[["body_mass_g", "bill_length_mm", "sex", "species"]]
```

```python
df
```

```python
df.columns = ["body_mass_g", "bill_length_mm", "gender", "species"]
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
ols_formula = "body_mass_g ~ bill_length_mm + C(gender) + C(species)"
```

```python
X_train = df[["bill_length_mm", "gender", "species"]]
y_train = df[["body_mass_g"]]
```

```python
ols_data = pd.concat([X_train, y_train], axis =1)
OLS = ols(formula = ols_formula, data = ols_data)
model = OLS.fit()
model.summary()
```

```python
# tÃ­nh tuyáº¿n tÃ­nh 
sns.pairplot(dfs)
```

```python
#tÃ­nh chuáº©n tÃ¡c
residuals = model.resid
fig, axes = plt.subplots(1,2,figsize=(8,4))
sns.histplot(residuals, ax=axes[0])
axes[0].set_xlabel("asdgfsa")
axes[0].set_title("gyahfgsdhjf")
sm.qqplot(residuals, line='s', ax = axes[1])
axes[1].set_title("hshfs")
plt.tight_layout()
plt.show()
```

```python
# tÃ­nh Ä‘á»“ng nháº¥t
fig = sns.scatterplot(x = model.fittedvalues, y = model.resid)
fig.set_xlabel("hgasdgf")
fig.set_ylabel("gvsagdf")
fig.set_title("shgfhs")
fig.axhline(0)
plt.show()
```
